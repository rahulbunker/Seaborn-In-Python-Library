```python
data=sns.load_dataset("penguins").head(20)
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex",style="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex",style="sex",size="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex",style="sex",size="sex"
               ,sizes=(100,400))
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex",style="sex",size="sex"
               ,sizes=(100,400),palette="Accent")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex",style="sex",size="sex"
               ,sizes=(100,400),palette="Accent",alpha=0.2)
# color light ho jayga
plt.show()
```

```python
m={"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex",style="sex",size="sex"
               ,sizes=(100,400),palette="Accent",markers=m)

plt.show()
```
