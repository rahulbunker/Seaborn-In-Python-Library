```python
data=sns.load_dataset("penguins")
data
```

```python
sns.displot(data["bill_length_mm"])
```

```python
sns.displot(data["bill_length_mm"])
plt.show()
 
```

```python
 sns.displot(data["flipper_length_mm"])
plt.show()

# esme data 170 -180 me shi nhi aa rha range me
 
```

```python
 sns.displot(data["flipper_length_mm"],bins=[170,180,190,200,210,220,230])
plt.show()
# esme perfect range show ho rhi hai
 
```

```python
sns.displot(data["flipper_length_mm"],bins=[170,180,190,200,210,220,230],kde=True)
plt.show()
# kde=>Kernel Density Estimate
```

```python
sns.displot(data["flipper_length_mm"],bins=[170,180,190,200,210,220,230],kde=True,rug=True)
plt.show()
```

```python
sns.displot(data["flipper_length_mm"],bins=[170,180,190,200,210,220,230],kde=True,rug=True,color="g")
plt.show()
```

```python
sns.displot(data["flipper_length_mm"],kde=True,rug=True,log_scale=True)
plt.show()
```
