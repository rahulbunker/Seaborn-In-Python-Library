# Bar Plot 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# df=sns.load_dataset(" ")
# sns.barplot(x=" ",y=" ",data=df)
# plt.show()
```

```python
df=sns.load_dataset("penguins")
df
```

```python
sns.barplot(x=df.island,y=df.bill_length_mm)
```

```python
sns.barplot(x=df.island,y=df.bill_length_mm)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex")
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",order=order_1)
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",order=order_1,
           hue_order=["Female","Male"]) # Female phle aajaygi
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",order=order_1,
           hue_order=["Female","Male"],ci=10)
# uper wala line ki size change hogi
plt.show()
```

```python

sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=df,hue="sex",orient="v")
# this is vertical
plt.show()
```

```python

sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=df,hue="sex",orient="h")
# this is horizental
plt.show()
```

```python

sns.barplot(x="island",y="bill_length_mm",data=df,color="g")

plt.show()
```

```python

sns.barplot(x="island",y="bill_length_mm",data=df,palette="Accent")

plt.show()
```

```python

sns.barplot(x="island",y="bill_length_mm",data=df,palette="icefire")

plt.show()
```

```python

sns.barplot(x="island",y="bill_length_mm",data=df,palette="Accent",saturation=0.1)

plt.show()
# color blur ho jaynge
```

```python

sns.barplot(x="island",y="bill_length_mm",data=df,palette="Accent",errcolor="y")

plt.show()
```

```python

sns.barplot(x="island",y="bill_length_mm",data=df,palette="Accent",errcolor="y",errwidth=10)

plt.show()
```

```python

sns.barplot(x="island",y="bill_length_mm",data=df,palette="Accent",errcolor="b",capsize=0.5)

plt.show()
```

```python

sns.barplot(x="island",y="bill_length_mm",data=df,palette="Accent",hue="sex",errcolor="y",dodge=False)

# esem male or female ek dusre par aajaynge
plt.show()
```

```python

sns.barplot(x="island",y="bill_length_mm",data=df,palette="Accent",hue="sex",errcolor="y",dodge=True)
# esme alg alg aaynge
plt.show()
```

```python

sns.barplot(x="island",y="bill_length_mm",data=df,palette="Accent",hue="sex",errcolor="y",dodge=False,
           alpha=0.1)
# graph ki brightness btata hai

plt.show()
```

```python
sns.set(style="dark")  # piche graph ke gridd aajygi
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",errcolor="y",dodge=False,
        )


plt.show()
```

```python
sns.set(style="darkgrid")  # piche graph ke gridd aajygi
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",errcolor="y",dodge=False,
        )


plt.show()
```
