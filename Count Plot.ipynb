# what is the diffrence between countplot and barplot ?

* Countplot plots the count of the number of records by category.
* Barplot plots a value or metric for each category(by default,barplot plots the mean of a variable,by category).

```python
df=sns.load_dataset("tips")
df
```

```python
sns.countplot(x="sex",data=df)
```

```python
sns.countplot(x="sex",data=df)
plt.show()
```

```python
sns.barplot(x="sex",y="size",data=df)
plt.show()
```

```python
sns.countplot(x="sex",data=df,hue="smoker")
plt.show()
```

```python
sns.countplot(y="sex",data=df,hue="smoker")
plt.show()
```

```python
sns.countplot(x="sex",data=df,hue="smoker",palette="bwr")
plt.show()
```

```python
sns.countplot(x="sex",data=df,hue="smoker",color="g",saturation=0.1)
# color ko light kar skte hai
plt.show()
```
